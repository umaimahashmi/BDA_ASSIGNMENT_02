```python
# Splitting words in SECTION_TEXT and combining into a single list
words_list = df['SECTION_TEXT'].str.split().tolist()
words_list_flat = [word for sublist in words_list for word in sublist]

# Creating a set to get unique words
unique_words_set = set(words_list_flat)

# Converting the set back to a list
vocabulary_list = sorted(list(unique_words_set))

print(vocabulary_list)
```

```python
print(len(vocabulary_list))
```

TF---> how often a word appears in a document

```python
#creating a  TF for all the documents
Term_Frequency_of_all_documents={}

for index, row in df.iterrows():
    empty_list = [0] * len(vocabulary_list)

    section_text = row['SECTION_TEXT']
    section_text_stripped = section_text.strip()
    words = section_text_stripped.split() #spaces excluded

    for word in words:
      word_to_find = word
      index_of_word = vocabulary_list.index(word_to_find)
      empty_list[index_of_word]+=1
    #print(empty_list)
    Term_Frequency_of_all_documents[index] = empty_list
# Assuming Term_Frequency_of_all_documents is your dictionary containing TF vectors

for document_index, tf_vector in Term_Frequency_of_all_documents.items():
    print(f"Document Index: {document_index}")
    print(f"TF Vector: {tf_vector}")
    print()  # Print an empty line for better readability
```

IDF  ---. INVERSE DOCUMENT FREQUENCY

```python

Inverse_Document_Frequency = [0] * len(vocabulary_list)

for word_index, check_word in enumerate(vocabulary_list):
    for index, row in df.iterrows():
        section_text = row['SECTION_TEXT']
        section_text_stripped = section_text.strip()
        words = section_text_stripped.split()  # Split the text into words

        if check_word in words:
            Inverse_Document_Frequency[word_index] += 1

print(Inverse_Document_Frequency)

```

```python
query = "as you sow , so shall you reap "
```

```python
def text_preprocess_query(query):
    # Lowercasing the query text
    query = query.lower()

    # Removing digits
    query = re.sub(r'\d+', '', query)

    # Removing special characters
    pattern = r'[^a-zA-Z0-9\s]'
    query = re.sub(pattern, '', query)

    # Splitting text on delimiters and removing punctuation
    tokens = re.findall(r'\b\w+\b', query)  # Using \b\w+\b to extract whole words
    tokens = [token.strip(string.punctuation) for token in tokens]

    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatizing tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Spell checking
    spell = SpellChecker()
    corrected_tokens = [spell.correction(token) for token in tokens]

    return " ".join(corrected_tokens)

query = text_preprocess_query(query)
print(query)
```

```python
# Calculate TF-IDF vector for the query "political philosophy"
def vector_for_query(query,vocabulary_list):
  query_words = query.split()
  query_tf_vector = [0] * len(vocabulary_list)

  for word in query_words:
    if word in vocabulary_list:
        word_index = vocabulary_list.index(word)
        query_tf_vector[word_index] += 1

  # Calculate TF-IDF values for the query
  for index, value in enumerate(query_tf_vector):
    if value != 0:
        query_tf_vector[index] = query_tf_vector[index] / Inverse_Document_Frequency[index]

  return query_tf_vector

query_tf_vector=vector_for_query(query,vocabulary_list)
print("TF-IDF Vector:")
print(query_tf_vector)
```

```python
import numpy as np
sum_values_all_documents = {}
for document_index, tf_vector in Term_Frequency_of_all_documents.items():
    sum = np.dot(tf_vector, query_tf_vector)  # Dot product of TF vector and query TF-IDF vector
    sum_values_all_documents[document_index] = sum

print("Cosine Similarity with Query:")
print(sum_values_all_documents)
```

```python
# Sort the documents by cosine similarity scores in descending order
sorted_documents = sorted(sum_values_all_documents.items(), key=lambda x: x[1], reverse=True)

# Print the top 5 documents with highest cosine similarity scores
print("Top 10 Documents with Highest Cosine Similarity:")
for index, (document_index, similarity_score) in enumerate(sorted_documents[:10], 1):
    print(f"Rank {index}: Document Index {document_index} - Cosine Similarity Score: {similarity_score}")
```
